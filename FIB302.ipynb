1 - Suponha que a posição de um ponto no espaço bidimensional é dado em coordenadas polares r, θ queremos convertê-lo em coordenadas cartesianas x , y. Escreva um programa para fazer essa conversão.

```python
import math

print("* Conversor de Coordenadas Polares para Cartesianas *")
r = float(input("Digite o valor do raio (r): "))
theta_graus = float(input("Digite o ângulo (theta) em graus: "))

# Conversão de graus para radianos
theta_rad = math.radians(theta_graus)

# Fórmulas: x = r * cos(theta), y = r * sin(theta)
x = r * math.cos(theta_rad)
y = r * math.sin(theta_rad)

print(f"\nResultados:")
print(f"Coordenada x: {x:.4f}")
print(f"Coordenada y: {y:.4f}")
```

2 - Escreva um programa que compute os principais parametros de um lançamento obliquo dado o módulo da velocidade inicial e o angulo de lançamento. Imprima a saída de maneira que os resultados sejam descritos claramente.

```python
import math

print("* Calculadora de Lançamento Oblíquo *")
v0 = float(input("Velocidade inicial (m/s): "))
angulo_graus = float(input("Ângulo de lançamento (graus): "))
g = 9.8  # Aceleração da gravidade

# Convertendo para radianos
theta = math.radians(angulo_graus)

# Tempo total de voo: T = 2 * v0 * sin(theta) / g
tempo_total = (2 * v0 * math.sin(theta)) / g

# Altura máxima: H = (v0^2 * sin^2(theta)) / 2g
altura_max = (v0**2 * math.sin(theta)**2) / (2 * g)

# Alcance horizontal: R = (v0^2 * sin(2*theta)) / g
alcance = (v0**2 * math.sin(2 * theta)) / g

print("\n* Parâmetros do Movimento *")
print(f"Tempo total de voo: {tempo_total:.2f} s")
print(f"Altura máxima:      {altura_max:.2f} m")
print(f"Distância (horizontal) percorrida: {alcance:.2f} m")
```

9. Defina arrays para velocidade inicial, ângulo e tempo. Calcule e imprima as posições horizontal e vertical de um projétil usando as equações de movimento pertinentes. Suponha que v_0 = 20m/s, theta = 45 graus e g = 9,8m/s2.

```python
# Movimento de Projétil
# Parâmetros
v0 = 20.0       # m/s
theta = 45.0    # graus
g = 9.8         # m/s^2

# Convertendo ângulo para radianos
theta_rad = np.deg2rad(theta)

# Criando um array de tempo (de 0 a 3 segundos, com 10 pontos)
t = np.linspace(0, 3, 10)

# Equações de movimento
# x = v0 * cos(theta) * t
pos_x = v0 * np.cos(theta_rad) * t
# y = v0 * sin(theta) * t - 0.5 * g * t^2
pos_y = v0 * np.sin(theta_rad) * t - 0.5 * g * t**2

print("* Projétil *")
print(f"Tempos: {t}")
print(f"Posições X: {pos_x}")
print(f"Posições Y: {pos_y}")
```

10. Defina arrays para tempo e deslocamento. Calcule e represente graficamente o deslocamento de um objeto em movimento harmônico simples. Suponha que a amplitude seja A = 1m e  a velocidade angular omega = 2*pi rad/s.

```python
import matplotlib.pyplot as plt

# Movimento Harmônico Simples (MHS)
A = 1.0         # Amplitude (m)
omega = 2 * np.pi # Velocidade angular (rad/s)
t_mhs = np.linspace(0, 2, 100) # 2 segundos

# x(t) = A * cos(omega * t)
deslocamento = A * np.cos(omega * t_mhs)

# Gráfico
plt.figure(figsize=(8, 4))
plt.plot(t_mhs, deslocamento)
plt.title("Movimento Harmônico Simples")
plt.xlabel("Tempo (s)")
plt.ylabel("Deslocamento (m)")
plt.grid(True)
plt.show()
```

1 - Faça um programa para calcular a sequencia de Fibonacci até 1000.

```python
print("* Sequência de Fibonacci (até 1000) *")

# Termos iniciais
a, b = 0, 1

# Enquanto 'a' for menor ou igual a 1000, execute o bloco
while a <= 1000:
    print(a, end=' ') # O end=' ' imprime na mesma linha

    # Atualização das variáveis:
    # 'a' recebe o valor de 'b'
    # 'b' recebe a soma antiga (a + b)
    a, b = b, a + b

print() # Apenas para pular uma linha no final
```

2 - Usando o inicio de código abaixo, finalize o programa para encontrar o tempo em que o projetil passa de sua altura máxima:

```python
from numpy import linspace

v0 = 4.5 # velocidade inicial
g = 9.81 # aceleração da gravidade
t = linspace(0, 1, 1000) # 1000 pontos do intervalo de tempo
y = v0*t - 0.5*g*t**2 # calcula todas as alturas do lançamento

# ache quando a bola passa de y=0

# Encontrar o índice (posição no array) onde y é máximo
indice_do_maximo = argmax(y)

# Usar esse índice para pegar o tempo correspondente no array t
tempo_max = t[indice_do_maximo]
altura_max = y[indice_do_maximo]

print(f"A altura máxima é atingida em t = {tempo_max:.4f} s")
print(f"A altura máxima é: {altura_max:.4f} m")
```

2 - Faça um programa que calcule as primeiras 5 linhas das 3 principais séries de linhas do Hidrogenio (m=1,2,3), Lyman, Balmer e Paschen de acordo com a fórmula de Rydberg

```python
# Constante de Rydberg (em nm^-1)
R = 1.097e-2

# Séries
series_nomes = {
    1: "Lyman (UV)",
    2: "Balmer (Visível)",
    3: "Paschen (Infravermelho)"
}

print(f"{'Série':<20} | {'Transição (n->m)':<18} | {'Comp. de Onda (nm)':<20}")
print("=" * 65)

# Loop pelas séries m=1, m=2, m=3
for m in [1, 2, 3]:

    # Loop pelas 5 primeiras linhas de cada série
    # As linhas começam sempre em n = m + 1
    for k in range(1, 6):
        n = m + k

        # Fórmula de Rydberg: 1/lambda = R * (1/m² - 1/n²)
        um_sobre_lambda = R * ( (1/(m**2)) - (1/(n**2)) )

        # Invertendo para achar o lambda
        lambda_nm = 1 / um_sobre_lambda

        # Imprimindo formatado
        nome = series_nomes[m] if k == 1 else "" # Mostra o nome apenas na 1ª linha do bloco
        print(f"{nome:<20} | n={n} -> m={m}      | {lambda_nm:.2f} nm")

    print("-" * 65) # Separador entre séries
```

3 - Escreva um programa que gere um número aleatório entre 1 e 100. Peça ao usuário para adivinhar o número. Continue avisando o usuário até que ele adivinhe corretamente. Forneça feedback se a estimativa for muito alta ou muito baixa.

```python
import random

print("* Adivinhe o número (1 a 100) *")
numero_secreto = random.randint(1, 100)
palpite = 0
tentativas = 0

# O loop continua enquanto o palpite for diferente do segredo
while palpite != numero_secreto:
    try:
        palpite = int(input("Seu palpite: "))
        tentativas += 1

        if palpite < numero_secreto:
            print("Muito baixo! Tente um número maior.")
        elif palpite > numero_secreto:
            print("Muito alto! Tente um número menor.")
        else:
            print(f"Parabéns! Você acertou o número {numero_secreto} em {tentativas} tentativas.")

    except ValueError:
        print("Por favor, digite um número inteiro válido.")
```

4 - Escreva um programa que peça ao usuário um número inteiro positivo e calcule seu fatorial usando um loop while.

```python
print("* Cálculo de Fatorial *")
n = int(input("Digite um número inteiro positivo: "))
original_n = n

fatorial = 1

if n < 0:
    print("Fatorial não existe para números negativos.")
elif n == 0:
    print("O fatorial de 0 é 1")
else:
    while n > 0:
        fatorial *= n  # Multiplica o acumulador pelo n atual
        n -= 1         # Decrementa n

    print(f"O fatorial de {original_n} é: {fatorial}")
```

6 - Escreva um programa que peça um número ao usuário e verifique se é um número primo.

```python
num = int(input("Digite um número para verificar se é primo: "))

eh_primo = True

if num <= 1:
    eh_primo = False
else:
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            eh_primo = False
            break # Interrompe o loop imediatamente se encontrar um divisor

if eh_primo:
    print(f"{num} é um número PRIMO.")
else:
    print(f"{num} NÃO é primo.")
```
